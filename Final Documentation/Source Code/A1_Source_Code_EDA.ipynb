## **Kaggle Binary**

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
```

```python
dataframe=pd.read_csv("Kaggle_Binary.csv")
df = pd.DataFrame(dataframe)
```

**Describing the data** 

```python
df.describe()
```

**First 5 rows**

```python
df.head()
```

**Boxplot**

```python
#df = df.drop(['is_safe'], axis=1)

df1=df.select_dtypes(exclude=['object'])
for column in df1:
        plt.figure(figsize=(17,1))
        sns.boxplot(data=df1, x=column)
```

**Heatmap**

```python
plt.figure(figsize=(18, 6))

heatmap = sns.heatmap(df.corr(),vmin=-1, vmax=1,annot=True,cmap="Blues")
heatmap.set_title('Binary-Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

**Class Distribution**

```python
ax=sns.countplot(df['is_safe'])

plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class distribution')
for p in ax.patches:
        ax.annotate('{:}'.format(p.get_height()), (p.get_x()+0.35, p.get_height()+50))
plt.show()
```

**Histogram**

```python

for column in df:
  plt.figure(figsize=(3,4))
  plt.hist(df[column], bins=9, color="skyblue",align='right', edgecolor='black')
  plt.xlabel(column)
  plt.grid(axis = 'y')
  plt.ylabel("Count")

plt.show()
```

**Kernel Density Estimation (KDE)**

```python
for column in df:
  plt.xlabel(column)
  plt.grid(axis = 'x')
  plt.grid(axis = 'y')
  sns.distplot(df[column],bins=55,fit_kws={"color":"skyblue"},kde_kws={"color": "k", "lw": 1, "label": "KDE"})

  plt.figure(figsize=(3,4))
```
